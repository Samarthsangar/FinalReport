```python
#taking order id and AWS number 
order_id= Invoice[['Order ID']]
AWB_Number= Invoice[['AWB Code']]
```

# Claculations

```python
Total_weight_x=SKU_Master['Weight (g)']/1000
Total_weight_CourierCompany=Invoice['Charged Weight']/1000
```

```python
Weight_slab_X =Total_weight_x/2.5
Weight_slab_CourierCompany = Total_weight_CourierCompany/2.5
```

```python
Delivery_Zone_per_X =Invoice[['Zone']]
Delivery_Zone_charged_by_Courier_Company=Invoice[['Charged Weight']]
```

```python
# Assuming Rs. 500 per KG
Expected_Charge_as_per_X = Total_weight_x*Invoice['Charged Weight']*500
```

```python
Charges_Billed_by_Courier_Company =Invoice[['Billing Amount (Rs.)']]
```

# generating new columns as per requrements
